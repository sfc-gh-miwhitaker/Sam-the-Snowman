# Chapter 9 - Teardown and Take It Home

Clean up demo objects and leave with a reusable design checklist.

## Learn
- How to safely remove walkthrough artifacts
- How to transfer this pattern to your own domain


```sql
-- Teardown
USE ROLE ACCOUNTADMIN;
USE WAREHOUSE SFE_SAM_SNOWMAN_WH;

DROP AGENT IF EXISTS SNOWFLAKE_EXAMPLE.SAM_DRIFT.SAM_THE_SNOWMAN;
DROP SEMANTIC VIEW IF EXISTS SNOWFLAKE_EXAMPLE.SEMANTIC_MODELS.SV_SAM_DRIFT_NAIVE;
DROP SEMANTIC VIEW IF EXISTS SNOWFLAKE_EXAMPLE.SEMANTIC_MODELS.SV_SAM_DRIFT_BASE;
DROP SEMANTIC VIEW IF EXISTS SNOWFLAKE_EXAMPLE.SEMANTIC_MODELS.SV_SAM_DRIFT_ONTOLOGY;

DROP SCHEMA IF EXISTS SNOWFLAKE_EXAMPLE.SAM_DRIFT;

-- Optional full cleanup
-- DROP GIT REPOSITORY IF EXISTS SNOWFLAKE_EXAMPLE.GIT_REPOS.SFE_SAM_THE_SNOWMAN_REPO;
-- DROP WAREHOUSE IF EXISTS SFE_SAM_SNOWMAN_WH;
```

## Take-Home Checklist
- What are your domain's `Person`, `Asset`, and `Event` abstractions?
- Which metrics need explicit formulas to avoid ambiguity?
- Which five hard questions should your agent pass before launch?
- Which tool should answer each hard question?
- What is your evaluation rubric and release gate?
