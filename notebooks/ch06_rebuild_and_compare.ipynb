# Chapter 6 - Rebuild Sam and Compare

Rebuild the agent using curated tools and rerun the same hard-question set.

## Learn
- How explicit routing improves answer quality
- How to quantify improvement versus chapter 2 baseline


```sql
-- Rebuild Sam with two curated semantic tools
CREATE OR REPLACE AGENT SNOWFLAKE_EXAMPLE.SAM_DRIFT.SAM_THE_SNOWMAN
  COMMENT = 'DEMO: Sam Drift ontology agent (Expires: 2026-04-18)'
  PROFILE = '{"display_name": "Sam-the-Snowman", "color": "blue"}'
  FROM SPECIFICATION
$$
models:
  orchestration: auto

instructions:
  system: |
    You are Sam, the in-house data snowman for Drift Entertainment.
  response: |
    Start with a direct answer, show calculations, then recommend next actions.

tools:
  - tool_spec:
      type: cortex_analyst_text_to_sql
      name: drift_base_query
      description: Use for precise source-table metrics, invoices, and support rep revenue.
    tool_resources:
      semantic_view: SNOWFLAKE_EXAMPLE.SEMANTIC_MODELS.SV_SAM_DRIFT_BASE
      execution_environment:
        warehouse: SFE_SAM_SNOWMAN_WH

  - tool_spec:
      type: cortex_analyst_text_to_sql
      name: drift_ontology_query
      description: Use for cross-entity or abstraction questions spanning people/media/sales.
    tool_resources:
      semantic_view: SNOWFLAKE_EXAMPLE.SEMANTIC_MODELS.SV_SAM_DRIFT_ONTOLOGY
      execution_environment:
        warehouse: SFE_SAM_SNOWMAN_WH
$$;
```
