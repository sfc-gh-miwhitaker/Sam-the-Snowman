# Chapter 3 - Think in Entities, Not Tables

Design ontology concepts before writing more SQL.

## Learn
- Why abstract parents improve routing and interpretation
- How to map Drift tables into reusable concepts


```mermaid
flowchart TD
  Person[Person Abstract] --> Customer[Customer]
  Person --> Employee[Employee]
  MediaEntity[MediaEntity Abstract] --> Artist[Artist]
  MediaEntity --> Album[Album]
  MediaEntity --> Track[Track]
  Sale[Sale Abstract] --> Invoice[Invoice]
  Sale --> InvoiceLine[InvoiceLine]
  Customer -->|purchased| Invoice
  Employee -->|supports| Customer
  Track -->|curated_on| Playlist[Playlist]
```


## Mapping Table
| Ontology concept | Source objects |
|---|---|
| `Person` | `CUSTOMER`, `EMPLOYEE` |
| `MediaEntity` | `ARTIST`, `ALBUM`, `TRACK` |
| `Sale` | `INVOICE`, `INVOICE_LINE` |
| `supports` | `CUSTOMER.SUPPORT_REP_ID -> EMPLOYEE.EMPLOYEE_ID` |
| `curated_on` | `PLAYLIST_TRACK` |


````sql
-- Check assumptions quickly
SELECT COUNT(*) AS customers FROM SNOWFLAKE_EXAMPLE.SAM_DRIFT.CUSTOMER;
SELECT COUNT(*) AS employees FROM SNOWFLAKE_EXAMPLE.SAM_DRIFT.EMPLOYEE;
SELECT MIN(INVOICE_DATE) AS min_dt, MAX(INVOICE_DATE) AS max_dt FROM SNOWFLAKE_EXAMPLE.SAM_DRIFT.INVOICE;
````
